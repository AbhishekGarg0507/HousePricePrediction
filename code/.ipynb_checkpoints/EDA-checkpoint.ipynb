```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('../data/real_estate_listings_clean.csv')
```

```python
print("Missing 'energy':", (df['energy'] == 'na').sum())
print("Missing 'heating':", (df['heating'] == 'na').sum())
```

```python
df['energy'].value_counts()
```

```python
df['heating'].value_counts()
```

```python
df['energy'] = df['energy'].replace('na', 'Unknown')
df['heating'] = df['heating'].replace('na', 'Unknown')
```

```python
# Remove trailing commas and whitespace
df['energy'] = df['energy'].str.strip().str.rstrip(',')
df['heating'] = df['heating'].str.strip().str.rstrip(',')

# Group rare categories (less than 20 occurrences) into 'Other'
energy_counts = df['energy'].value_counts()
rare_energies = energy_counts[energy_counts < 20].index
df['energy'] = df['energy'].replace(rare_energies, 'Other')

heating_counts = df['heating'].value_counts()
rare_heatings = heating_counts[heating_counts < 20].index
df['heating'] = df['heating'].replace(rare_heatings, 'Other')
```

```python


df['energy'].value_counts().plot(kind='bar', figsize=(10,4), title='Energy Types')
plt.show()

df['heating'].value_counts().plot(kind='bar', figsize=(10,4), title='Heating Types')
plt.show()
```

```python
df.max()
```

```python
df.min()
```

```python
# Cap price at 99th percentile
price_cap = df['price'].quantile(0.99)
df = df[df['price'] <= price_cap]

# Cap area at 99th percentile
area_cap = df['area'].quantile(0.99)
df = df[df['area'] <= area_cap]

# Cap rooms at 10 (rare to have more in residential)
df = df[df['rooms'] <= 10]

# Cap fee at 99th percentile
fee_cap = df['fee'].quantile(0.99)
df = df[df['fee'] <= fee_cap]

# Cap price_per_area at 99th percentile
ppa_cap = df['price_per_area'].quantile(0.99)
df = df[df['price_per_area'] <= ppa_cap]
```

```python


# Fix construction_year
df.loc[(df['construction_year'] < 1800) | (df['construction_year'] > 2025), 'construction_year'] = np.nan

# Fix level
df.loc[(df['level'] < 0) | (df['level'] > 20), 'level'] = np.nan
```

```python
df['log_price'] = np.log1p(df['price'])
df['log_area'] = np.log1p(df['area'])
df['log_price_per_area'] = np.log1p(df['price_per_area'])
```

```python
df.describe()
```

```python

for col in ['price', 'area', 'rooms', 'fee', 'price_per_area', 'log_price', 'log_area', 'log_price_per_area']:
    plt.figure(figsize=(6, 4))
    df[col].hist(bins=50)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
numeric_features = ['price', 'area', 'rooms', 'fee', 'price_per_area', 'log_price', 'log_area', 'log_price_per_area']

for feature in numeric_features:
    plt.figure(figsize=(6, 1.5))
    plt.boxplot(df[feature].dropna(), vert=False)
    plt.title(f'Boxplot of {feature}')
    plt.xlabel(feature)
    plt.show()
```
